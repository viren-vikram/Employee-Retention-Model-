```python
# print_function for compatibility with Python 3
from __future__ import print_function

# NumPy for numerical computing
import numpy as np
# Pandas for DataFrames
import pandas as pd
pd.set_option('display.max_columns',100)
# Matplotlib for visualization
from matplotlib import pyplot as plt
# display plots in the notebook
%matplotlib inline
# Seaborn for easier visualization
import seaborn as sns
```

```python
# Load employee data from CSV
df=pd.read_csv('project_files/employee_data.csv')
```

```python
# Plot histogram grid

df.hist(figsize=(10,10),xrot=-45)
# Clear the text "residue"
plt.show()
```

```python
# Summarize numerical features
df.describe()
```

```python
# Summarize categorical features
df.describe(include=['object'])
```

```python
# Plot bar plot for each categorical feature
#sns.countplot(y='department',data=df)
#plt.show()
#sns.countplot(y='salary',data=df)
#plt.show()
#sns.countplot(y='status',data=df)
#plt.show()
#
for feature in df.dtypes[df.dtypes=='object'].index:
    sns.countplot(y=feature,data=df)
    plt.show()
```

```python
# Segment satisfaction by status and plot distributions
sns.violinplot(y='status',x='satisfaction',data=df)
plt.show()
```

```python
# Segment last_evaluation by status and plot distributions
sns.violinplot(y='status',x='last_evaluation',data=df)
plt.show()
```

```python
# Segment by status and display the means within each class
df.groupby('status').mean()
```

```python
correlations=df.corr()
```

```python
#set color theme
sns.set_style('white')
#set figure size
plt.figure(figsize=(8,8))
#set mask for removing one side of the triangle
mask=np.zeros_like(correlations,dtype=np.bool)
mask[np.triu_indices_from(mask)]=True
sns.heatmap(correlations*100,mask=mask,annot=True,fmt='.0f',cbar=False)
plt.show()
```

```python
# Scatterplot of satisfaction vs. last_evaluation
sns.lmplot(y='last_evaluation',x='satisfaction',hue='status',data=df,fit_reg=False)
plt.show()
```

```python
# Scatterplot of satisfaction vs. last_evaluation, only those who have left
sns.lmplot(x='satisfaction',y='last_evaluation',data=df[df['status']=='Left'],fit_reg=False)
plt.show()
```
